```python
import pandas as pd
import numpy as np


```

```python
file_path = "Meta_1_Finito.xlsx"

# Caricare il file Excel per esaminare il contenuto
xls = pd.ExcelFile(file_path)
xls.sheet_names
df = pd.read_excel(xls, sheet_name="Meta")
```

```python


# Definizione della nuova tabella con colonne analoghe all'immagine
new_df = pd.DataFrame()

# Estrarre e adattare i dati
new_df["ID_cantiere"] = df["ID_Cantiere"]  # Rimuovere l'underscore per uniformità
new_df["Zona"] = df["Inserzioni"].str.split("|").str[0].str.strip()
new_df["Nome Cantiere"] = df["Inserzioni"].str.split("|").str[1].str.strip()
new_df["Città"] = df["Inserzioni"].str.split("|").str[2].str.strip()
new_df["Tipologia"] = np.random.choice(["Palazzo", "Condominio",'Villette'], size=len(df))  # Scelta casuale tra tipologie
new_df["Garage"] = np.random.choice(["Sì", "No"], size=len(df))   
new_df["Ascensore"] = np.random.choice(["Sì", "No"], size=len(df))  # Valori casuali
new_df["Giardino"] = np.random.choice(["Sì", "No"], size=len(df))
new_df["Posizione"] = np.random.choice(["Centro", "Periferia"], size=len(df))  # Valori casuali
new_df['Unità'] = np.random.randint(6, 25, size=len(df))
new_df["M² Med"] = np.random.randint(80, 150, size=len(df))  # Metri quadrati medi
new_df["M² Totali"] = new_df["M² Med"] * new_df['Unità']  # Totale metri quadrati
new_df["Classe En."] = np.random.choice(["A1", "A2", "A3",'A4','B','C'], size=len(df))  # Classi energetiche casuali
new_df["Prezzo/M²"] = np.random.randint(2000, 5000, size=len(df))  # Prezzo al metro quadro
new_df["Prezzo Unità"] = new_df["M² Med"] * new_df["Prezzo/M²"]  # Prezzo per unità
new_df["Prezzo Totale"] = new_df["M² Totali"] * new_df["Prezzo/M²"]  # Prezzo totale

# Visualizzare le prime righe della tabella generata
new_df.head()
```

```python
# Percorso per il file da esportare
output_file_path = "Tabella_Cantieri.xlsx"

# Esportare il DataFrame in un file Excel
new_df.to_excel(output_file_path, index=False)

# Restituire il percorso del file generato
output_file_path
```
